```python
# import basice apis
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from skgarden.mondrian.ensemble import MondrianForestClassifier

# import Sample Data to learn models
dataset = load_breast_cancer()
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.DataFrame(dataset.target, columns=['y'])

# split data for crossvalidation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)



# set pipelines for different algorithms
pipe_rf = Pipeline([('scl',StandardScaler()),('est',RandomForestClassifier(random_state=1))])
pipe_gb = Pipeline([('scl',StandardScaler()),('est',GradientBoostingClassifier(random_state=1))])
pipe_mf = Pipeline([('scl',StandardScaler()),('est',MondrianForestClassifier())])


# fit & evaluation
pipe_rf.fit(X_train,y_train.as_matrix().ravel()) 
pipe_gb.fit(X_train,y_train.as_matrix().ravel()) 
pipe_mf.fit(X_train,y_train.as_matrix().ravel())

def variable_score():
    

print('RandomForest:%.3f' %accuracy_score(y_test.as_matrix().ravel(), pipe_rf.predict(X_test)))
print('GradientBoosting:%.3f' %accuracy_score(y_test.as_matrix().ravel(), pipe_gb.predict(X_test)))
print('MondrianForest:%.3f' %accuracy_score(y_test.as_matrix().ravel(), pipe_mf.predict(X_test)))
```

```python
#データセットを5個に分ける

X1 = X_train.iloc[:150,:]
y1 = y_train.iloc[:150,:]
X2 = X_train.iloc[150:200,:]
y2 = y_train.iloc[150:200,:]
X3 = X_train.iloc[200:250,:]
y3 = y_train.iloc[200:250,:]
X4 = X_train.iloc[250:300,:]
y4 = y_train.iloc[250:300,:]
X5 = X_train.iloc[300:350,:]
y5 = y_train.iloc[300:350,:]
X6 = X_train.iloc[350:400,:]
y6 = y_train.iloc[350:400,:]
X7 = X_train.iloc[400:450,:]
y7 = y_train.iloc[400:450,:]

print('X1 shape: (%i,%i)' %X1.shape)
print('y1 shape: (%i,%i)' %y1.shape)
print('----------------------------------------------------------------------------------------')
print('X3 shape: (%i,%i)' %X3.shape)
print('y3 shape: (%i,%i)' %y3.shape)
print('----------------------------------------------------------------------------------------')
print('----------------------------------------------------------------------------------------')
print('X5 shape: (%i,%i)' %X5.shape)
print('y5 shape: (%i,%i)' %y5.shape)
print('----------------------------------------------------------------------------------------')
print('----------------------------------------------------------------------------------------')
print('X6 shape: (%i,%i)' %X6.shape)
print('y6 shape: (%i,%i)' %y6.shape)

print('----------------------------------------------------------------------------------------')
print('----------------------------------------------------------------------------------------')
print('X7 shape: (%i,%i)' %X7.shape)
print('y7 shape: (%i,%i)' %y7.shape)

print('----------------------------------------------------------------------------------------')
```

```python
from skgarden.mondrian.ensemble import MondrianForestRegressor
from sklearn.preprocessing import StandardScaler

# build the model
scaler = StandardScaler()
scaler.fit(X1)
X1_scl = scaler.transform(X1)
mfr = MondrianForestClassifier()
mfr.fit(X1_scl,y1.as_matrix().ravel())

ac1 = accuracy_score(y_test.as_matrix().ravel(), mfr.predict(X_test))
print('accuracy score of the MondrianForest model: %.3f' % ac1)
```
