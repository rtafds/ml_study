# WORK for P01

## [Work-1]

- 機械学習とは何か？
    - 人間の介在なく機械がデータから学び判断・振る舞いを決められるようにする技術・研究領域
<br><br>
- 機械学習を分類するとどうなるか？
    - 教師あり学習、教師なし学習、強化学習
<br><br>
- 教師あり学習と教師なし学習の違いは何か？
    - 教師あり学習は正確データを持つが教師なし学習にはない。
    - 教師あり学習は特徴量と正解データの相関関係の学習から主に予測に使われる。
    - 教師なし学習は特徴量データ間の構造把握などに使われる。
<br><br>
- 精度の高いモデル構築には何が大切か？
    - 多種多様なデータができれば大量に(多様性の例:長い時間軸、データ収集ポイントのエリア、顧客属性など)
<br><br>
- AML(Automated Machine Learninig)とはどのようなものか？
    - アルゴリズムがもつハイパーパラメータ(損失関数の最適化対象外のパラメータ)までも最適化対象とした機械学習の枠組み
